```python
df_country =pd.read_excel("/content/Country-Code.xlsx")
df_country.head()
```

```python
df_country.columns
```

```python
final_df=pd.merge(df,df_country,on="Country Code",how='left')
final_df.head(2)
```

```python
final_df.info()
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#Pie chart
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.1f%%')
```

Observations:

1 India is the highest consumer of the zomato

2 zomato has max transactions in india after that usa and then uk than so on

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"rating count"})
```

```python
ratings
```

Observation

1 When rating is 4.5 to 4.9 then it is excellent

2 when the rating is 4.0 to 4.4 then it is very good

3 when the rating is 3.5 to 3.9 then it is good

4 when the rating is 2.5 to 3.4 then it is avg

5 when the rating is 1.8 to 2.4 then it is poor

6 rating 0 then not rated

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='rating count',hue="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation

1-Not rated count is very high

2 Maximum number of rating are between 2.5 to 3.4



```python
sns.countplot(x='Rating color',data=ratings ,palette=['blue','red','orange','yellow','green','green'])
```
